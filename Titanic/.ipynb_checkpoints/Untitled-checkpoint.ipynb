```python
# Getting data ready
# Import needed moduls

import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

# Modelling
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split, GridSearchCV

# Setup random seed
np.random.seed(42)
```

```python
# Load data

titanic_train = pd.read_csv("data/train.csv", index_col='PassengerId')
titanic_test = pd.read_csv("data/test.csv", index_col='PassengerId')

titanic_train.head()
```

```python
not_needed_cols1 = []
not_needed_cols = ["Ticket", "Cabin", "Name"]
# # Split data
X = titanic_train.drop(no_need1 + ["Survived"], axis=1)
y = titanic_train["Survived"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train.head()
```

```python
mask = (X_train.dtypes == 'object')

object_cols = list(mask[mask].index)
num_cols = list(mask[~mask].index)
```

```python
mask = (titanic_train.dtypes == 'object')

object_cols = list((set(mask[mask].index)-set(not_needed_cols)))
num_cols = list((set(mask[~mask].index)-set(not_needed_cols)))
```

```python
object_cols, num_cols
```

```python

# not_needed_transformer = Pipeline(steps=[
#     ("no_needed", no_needed_cols())])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="constant", fill_value="missing")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))])

numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median"))
])

    
preprocessor = ColumnTransformer(
    transformers=[        
        ("obj", categorical_transformer, object_cols),
        ("num", numeric_transformer, num_cols)])

model = Pipeline(steps=[("preprocessor", preprocessor),
                        ("model", RandomForestRegressor())])


# # Fit and score the model
model.fit(X_train, y_train)
model.score(X_test, y_test)
```
